```python
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
#data aquisition

with open("Data2.txt", "r") as f:
    Data = [[float(num) for num in line.split(' ')] for line in f]
    
data_transposed = Data
df = pd.DataFrame(data_transposed, columns=["Size", "Gauss_Algorithm", "Strassen_Algorithm", "Strassen_Algorithm_Memory_Efficient"])
#plot
myplot = sns.lineplot(data = df, x = "Size", y = "Gauss_Algorithm")
myplot = sns.lineplot(data = df, x = "Size", y = "Strassen_Algorithm")
myplot = sns.lineplot(data = df, x = "Size", y = "Strassen_Algorithm_Memory_Efficient")
myplot.set(xlabel='Size', ylabel='Time (s)')
plt.savefig('Plot1.png',dpi=300)
plt.close()

myplot1 = sns.lineplot(data = df, x = "Size", y = "Strassen_Algorithm")
myplot1 = sns.lineplot(data = df, x = "Size", y = "Strassen_Algorithm_Memory_Efficient")
myplot1.set(xlabel='Size', ylabel='Time (s)')
plt.savefig('Plot2.png',dpi=300)
```
